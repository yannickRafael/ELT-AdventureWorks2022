<h1>Carregar os dados<h1>

```python
DATASET_PATH = "./data/PRODUCTION_DATA.csv"
df = pd.read_csv(DATASET_PATH)
```

# Tratamento de dados

```python
def cor_para_decimal(cor):
    cor_hex = mcolors.CSS4_COLORS.get(cor.lower())
    if cor_hex:
        # Converter o valor hexadecimal para inteiro decimal
        return int(cor_hex.lstrip('#'), 16)  # Remover '#' antes de converter
    else:
        raise ValueError("Nome de cor inválido")


df['color'] = df['color'].apply(cor_para_decimal)

print(df.info())
pd.set_option('display.max_columns', None)
print(df.head(3))
```
